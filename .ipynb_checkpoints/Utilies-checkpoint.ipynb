# Convert image to binary

```python
from PIL import Image
import glob
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from skimage import io
from skimage.color import rgb2gray
import os
from natsort import natsorted

files = natsorted(glob.glob(r'I:\DVRPC\vector\labels\*.PNG'))
saved_path = r'I:\DVRPC\vector\labels\binary'

if not os.path.exists(saved_path):
    os.mkdir(saved_path)

def img_to_binary(img):    
    gray = io.imread(img)
    gray = gray > 0
    gray = gray.astype(np.uint8)     
    return gray

for file in files:
    print('Processing: ', file)
    img = img_to_binary(file)
    new_name = os.path.join(saved_path, os.path.basename(file))
    io.imsave(new_name, img, check_contrast=False)
```
