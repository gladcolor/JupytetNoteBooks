```python
 
from mmdet.apis import init_detector, inference_detector, show_result_pyplot, show_result_ins
import mmcv
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import os
import sys
import glob
```

```python
config_file = '/media/huan/HD4T/Research/SOLO/configs/vert/decoupled_vert_light_r50_fpn_8gpu_3x.py'
# download the checkpoint from model zoo and put it in `checkpoints/`
checkpoint_file = '/media/huan/HD4T/Research/SOLO/work_dirs/vert_light_release_r50_fpn_8gpu_3x/epoch_2.pth'
```

```python
# build the model from a config file and a checkpoint file
model = init_detector(config_file, checkpoint_file, device='cuda:0')
```

```python
# test a single image
img = '/media/huan/HD4T/Research/Park_face/images/atlanta beltline-20200521T115439Z-005/atlanta beltline/photos/beltline1036.jpg'
result = inference_detector(model, img)
```

```python
files = glob.glob(r'/media/huan/SSD/Datasets/AIRS/test/images/*.JPG')
save_path = r'/media/huan/SSD/Datasets/AIRS/test/inferenced2'
# os.mkdir(save_path)
len(files)
```

```python
# show the results
for idx, file in enumerate(files[252:]):
    try:
        bname = os.path.basename(file)
        img = file
        new_fullname = os.path.join(save_path, bname)
        result = inference_detector(model, img)
        show_result_ins(img, result, model.CLASSES, score_thr=0.25, out_file=new_fullname)
        if idx % 100 == 0:
            print(bname)
    except Exception as e:
        print("Error: ", e)
        continue
```
