```python
import glob
import PIL
import numpy as np
from PIL import Image
from natsort import natsorted
import matplotlib.pyplot as plt

folder = r'X:\My Drive\vacant_house\images\non_vacant_seg\*.png'
files = glob.glob(folder)

files2 = []

for f in files:
    if  "_color.png" in f:
        files2.append(f)
        
files2 = natsorted(files2)
```

```python
import shutil
import os
from tqdm import tqdm

image_dir = r'XXXXXXXXXXXXXXXXXXXXXXXXXXXXX'

for png in tqdm(files2[:]):
#     print(png)
    img = Image.open(png)
#     plt.imshow(img)
#     plt.show()
    
    img = np.array(img)
    
    h = 768
    w = 768
    
    clip_img = img[0:int(h/2), int(h/3):int(h*2/3)]
    
    hist = np.histogram(clip_img, range(153))
    
#     print(img)
#     print(hist[0])
#     print(hist[0].sum())
#     print("wall:", hist[0][1])
#     print("building:", hist[0][2])
#     print("tree:", hist[0][5])
#     print("grass:", hist[0][10])
#     print("plant:", hist[0][18])
#     print("house:", hist[0][26])
#     print("palm;palm;tree:", hist[0][73])
    
    house_cnt =  hist[0][1] + hist[0][2] + hist[0][26]
    tree_cnt =  hist[0][5] + hist[0][18] + hist[0][73]
    
    
    

    
    if tree_cnt > (house_cnt * 1.5):
        basename = os.path.basename(png)
        basename = basename.replace('_color.png', '.jpg')

        old_name = os.path.join(image_dir, basename)
        new_name = os.path.join(image_dir, "occluded", basename)
#         print("old_name : ", old_name)
#         print("New name: ", new_name)
#         print("house_cnt, tree_cnt: ", house_cnt,  ', ',  tree_cnt)
        shutil.move(old_name, new_name)
        
#         plt.imshow(clip_img)
#         plt.show()
        
```
