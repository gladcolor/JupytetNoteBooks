```python
df_temp = df.groupby('date').count()
df_temp
```

```python
dates = list(df_temp.index)
```

```python
temp_file = r'K:\OneDrive_NJIT\OneDrive - NJIT\Group_work\GIS-COUNTY\covid_19_state_level_reports\covid_19_state_level_reports\05-27-2020.csv'
df_template = pd.read_csv(temp_file)
df_template.loc[2, 'FIPS'] = 11
```

```python
df_template
```

```python
df_template['Confirmed'] = 0
df_template['Deaths'] = 0
df_template['Recovered'] = -1
df_template['Active'] = -1
df_template
```

```python
saved_path = r'K:\OneDrive_NJIT\OneDrive - NJIT\Group_work\GIS-COUNTY\day_by_day'

for idx, d in enumerate(dates[:]):
    
    df_temp = df[df['date'] == d]
    df_temp = df_temp.sort_values('fips')
    df_tofill = df_template.copy()
    df_tofill['Last_Update'] = d + ' 23:59'
    
    for idx2, row in df_temp.iterrows():
        fips = row['fips']
        cases = row['cases']
        deaths = row['deaths']
        
        df_tofill.loc[df_tofill['FIPS'] == fips, 'Confirmed'] = cases
        df_tofill.loc[df_tofill['FIPS'] == fips, 'Deaths'] = deaths
    
    new_date = d[-5:] + "_" + d[:4]    
    new_file = os.path.join(saved_path, new_date + '.csv')
    df_tofill = df_tofill.sort_values('FIPS')
    df_tofill.to_csv(new_file, index=False)
    
    
    print("Saved: ", new_date)
        
#     print(df_tofill)
```
