```python
jieba.load_userdict("O:\OneDrive_NJIT\OneDrive - NJIT\Courses\IS665\Project\wh_dict.csv")

def isAddr(addr_str):
    last_char = str(addr_str)[-1]
#     print(last_char)
    if (last_char == '号') or (last_char.isnumeric()):
        return True
    
    elif (addr_str[-2:] == '号线'):
        return False
        
    else:
        return False

seg_addr = []
POI['seg_addr'] = ''
POI['is_addr'] = ''
POI['merged'] = ''
for idx, row in POI.iterrows():
    addr = row['address']
    types = row['type'].replace(';', "|")
    
    name = row['name']
#     print("原地名：", name)
    p_pos = name.find('(')
#     print('p_pos:', p_pos)
    if p_pos > 0:
        name = name[:p_pos]
     
        
    seg_name_list = list(jieba.cut(name))
    name = "|".join(seg_name_list)
    
#     if len(seg_name_list) < 2: # only one word, then copy to two words.
#         seg_name_list = seg_name_list * 2
#         print(seg_name_list)
    
    m_pos = math.ceil(len(seg_name_list)/2)
    name1 = seg_name_list[:m_pos]
    name2 = seg_name_list[m_pos:]
#     print(seg_name_list, name1, name2)
    
    
    
    p_pos = addr.find('(')
#     print('p_pos:', p_pos)
    if p_pos > 0:
        addr = addr[:p_pos]
    
#     p_pos = addr.rfind('号')
# #     print('p_pos:', p_pos)
#     if p_pos > 0:
#         addr = addr[:p_pos+1]
    
    addr = addr.replace("附近",'')
    
    seg_list = jieba.cut(addr, cut_all=False)
    seg_list = "|".join(seg_list)

#     seg_addr.append(seg_list)
    merged = types + " " + "|".join(name1) + "," + seg_list + "," + "|".join(name2)
    
    POI.at[idx, 'seg_addr'] = seg_list
    POI.at[idx, 'is_addr'] = isAddr(addr)
    if row['address'][-2:] == '号线':
        POI.at[idx, 'is_addr'] = False
    POI.at[idx, 'merged'] = merged
    
#     if len(list(jieba.cut(row['name']))) < 2:
#         print(list(jieba.cut(row['name'])))
#         print(addr)
    
    if idx % 99 == 0:
        print("原地址：", addr)
        print("清理后：",addr) 
        print("是否是地址：", isAddr(addr))
        print('name:', name)
        print('types:', types)
        print('merged:', merged)
        print("分词后：", seg_list + '\n')  # 精确模式
        
        
    
print(POI['seg_addr'])
POI.to_csv(file.replace(".csv", "_分词后2.csv"), index=False)

# with open(file.replace(".csv", "_分词后.csv"), 'w') as f:
#     f.writelines('\n'.join(seg_addr))
```
